# PIE PLOT

#Pie plot of popularity according to year

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv(r'D:\DataScience\MyProjects\data.csv')
df.groupby(['Year']).sum().plot( kind='pie', y='Popularity',autopct='%1.0f%%')
plt.show()
```
